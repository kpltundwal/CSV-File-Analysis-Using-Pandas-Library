# Data Manipulation with Pandas


### Import Pandas

```python
import pandas as pd
```

### Pandas Data Manipulation:
It is the process of changing, organizing, or filtering data using the Pandas library in Python.

### Read CSV File

```python
df = pd.read_csv("Data Science.csv")
```

### Show top 5 Row

```python
df = pd.read_csv("Data Science.csv")
df.head()
```

###  Drop Null Values

```python
df = df.dropna()           # # Remove missing values from dataset
df.head()
```

### Check actual column names in the dataset

```python
df.columns
```

###  Add New Column – Double CGPA

```python
df['Double_CGPA'] = df['CGPA'] * 2         #  Add a new column 'Double_CGPA' by multiplying CGPA by 2

df.head()
```

### Rename Column – 'Phone_Number' to 'Contact'


```python
df.rename(columns={'Phone_Number': 'Contact'}, inplace=True)          #  Rename the column 'Phone_Number' to 'Contact'

df.head()
```

### Filter students with Age > 22

```python
filtered_df = df[df['Age'] > 22]                     # Filter students older than 22 years

filtered_df.head()
```

### Sort data by CGPA in descending order

```python
df = df.sort_values(by='CGPA', ascending=False)            # Cell 9: Sort dataframe by CGPA in descending order

df.head()
```

### Group by Department and calculate average CGPA

```python
grouped = df.groupby('Department')['CGPA'].mean()             #  Group data by Department and get average CGPA
grouped.head()
```
